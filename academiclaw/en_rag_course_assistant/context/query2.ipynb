## 1、Query描述
你需要实现一个基于 RAG的智能课程助教系统。系统从课程文档（PDF、PPTX、DOCX、TXT）中抽取文本，进行分块，生成向量并写入向量数据库（ChromaDB），随后在问答时先检索再生成答案，并在回答中标注引用来源（文件名与页码/幻灯片编号）。

你必须补全项目中标记为 [TODO] 的核心模块，使整个系统能够从零开始完成：建库 → 交互问答。

功能要求：
- 文档解析：支持 PDF、PPTX、DOCX、TXT 四种格式的文本提取；并为每个提取单元提供可追溯元信息（如页码、幻灯片页、段落序等）。
- 文本切分：将长文档切分成不超过 chunk_size 的块，并且相邻块之间保留 chunk_overlap 重叠。
- 向量库：调用 Embedding API 生成向量，写入 ChromaDB；实现 Top-K 相似检索。
- RAG Agent：
    - 定义 System Prompt（课程助教角色、引用规则、禁止编造）
    - 检索并格式化上下文（必须包含来源：文件名 + 页码/slide）
    - 调用 LLM 生成回答；回答必须“基于检索证据”，并带引用标注。
    - 可复现运行要求
    - 运行 python process_data.py：能完成数据处理与建库，不报错
    - 运行 python main.py：进入交互模式，输入问题能返回回答并带引用
    - data/ 中无论是否包含某种格式文件，你的 loader 都必须实现该格式的处理方法（即 4 个 loader 都要写）
